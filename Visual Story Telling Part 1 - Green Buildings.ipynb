The case

Over the past decade, both investors and the general public have paid increasingly close attention to the benefits of environmentally conscious buildings. There are both ethical and economic forces at work here. In commercial real estate, issues of eco-friendliness are intimately tied up with ordinary decisions about how to allocate capital. In this context, the decision to invest in eco-friendly buildings could pay off in at least four ways.

- Every building has the obvious list of recurring costs: water, climate control, lighting, waste disposal, and so forth. Almost by definition, these costs are lower in green buildings.
- Green buildings are often associated with better indoor environments—the kind that are full of sunlight, natural materials, and various other humane touches. Such environments, in turn, might result in higher employee productivity and lower absenteeism, and might therefore be more coveted by potential tenants. The financial impact of this factor, however, is rather hard to quantify ex ante; you cannot simply ask an engineer in the same way that you could ask a question such as, “How much are these solar panels likely to save on the power bill?”
- Green buildings make for good PR. They send a signal about social responsibility and ecological awareness, and might therefore command a premium from potential tenants who want their customers to associate them with these values. It is widely believed that a good corporate image may enable a firm to charge premium prices, to hire better talent, and to attract socially conscious investors.
- Finally, sustainable buildings might have longer economically valuable lives. For one thing, they are expected to last longer, in a direct physical sense. (One of the core concepts of the green-building movement is “life-cycle analysis,” which accounts for the high front-end environmental impact of ac- quiring materials and constructing a new building in the first place.) Moreover, green buildings may also be less susceptible to market risk—in particular, the risk that energy prices will spike, driving away tenants into the arms of bolder, greener investors.

Of course, much of this is mere conjecture. At the end of the day, tenants may or may not be willing to pay a premium for rental space in green buildings. We can only find out by carefully examining data on the commercial real-estate market.

The file greenbuildings.csv contains data on 7,894 commercial rental properties from across the United States. Of these, 685 properties have been awarded either LEED or EnergyStar certification as a green building. You can easily find out more about these rating systems on the web, e.g. at www.usgbc.org. The basic idea is that a commercial property can receive a green certification if its energy efficiency, carbon footprint, site selection, and building materials meet certain environmental benchmarks, as certified by outside engineers.

A group of real estate economists constructed the data in the following way. Of the 1,360 green-certified buildings listed as of December 2007 on the LEED or EnergyStar websites, current information about building characteristics and monthly rents were available for 685 of them. In order to provide a control population, each of these 685 buildings was matched to a cluster of nearby commercial buildings in the CoStar database. Each small cluster contains one green-certified building, and all non-rated buildings within a quarter-mile radius of the certified building. On average, each of the 685 clusters contains roughly 12 buildings, for a total of 7,894 data points.

The columns of the data set are coded as follows:

- CS.PropertyID: the building's unique identifier in the CoStar database.
- cluster: an identifier for the building cluster, with each cluster containing one green-certified building and at least one other non-green-certified building within a quarter-mile radius of the cluster center.
- size: the total square footage of available rental space in the building.
- empl.gr: the year-on-year growth rate in employment in the building's geographic region.
- Rent: the rent charged to tenants in the building, in dollars per square foot per calendar year.
- leasing.rate: a measure of occupancy; the fraction of the building's available space currently under lease.
- stories: the height of the building in stories.
- age: the age of the building in years.
- renovated: whether the building has undergone substantial renovations during its lifetime.
- class.a, class.b: indicators for two classes of building quality (the third is Class C). These are relative classifications within a specific market. Class A buildings are generally the highest-quality properties in a given market. Class B buildings are a notch down, but still of reasonable quality. Class C buildings are the least desirable properties in a given market.
- green.rating: an indicator for whether the building is either LEED- or EnergyStar-certified.
- LEED, Energystar: indicators for the two specific kinds of green certifications.
- net: an indicator as to whether the rent is quoted on a "net contract" basis. Tenants with net-rental contracts pay their own utility costs, which are otherwise included in the quoted rental price.
- amenities: an indicator of whether at least one of the following amenities is available on-site: bank, convenience store, dry cleaner, restaurant, retail shops, fitness center.
- cd.total.07: number of cooling degree days in the building's region in 2007. A degree day is a measure of demand for energy; higher values mean greater demand. Cooling degree days are measured relative to a baseline outdoor temperature, below which a building needs no cooling.
- hd.total07: number of heating degree days in the building's region in 2007. Heating degree days are also measured relative to a baseline outdoor temperature, above which a building needs no heating.
- total.dd.07: the total number of degree days (either heating or cooling) in the building's region in 2007.
- Precipitation: annual precipitation in inches in the building's geographic region.
- Gas.Costs: a measure of how much natural gas costs in the building's geographic region.
- Electricity.Costs: a measure of how much electricity costs in the building's geographic region.
- cluster.rent: a measure of average rent per square-foot per calendar year in the building's local market.

The goal

An Austin real-estate developer is interested in the possible economic impact of "going green" in her latest project: a new 15-story mixed-use building on East Cesar Chavez, just across I-35 from downtown. Will investing in a green building be worth it, from an economic perspective? The baseline construction costs are $100 million, with a 5 percent expected premium for green certification.

The developer has had someone on her staff, who's been described to her as a "total Excel guru from his undergrad statistics course," run some numbers on this data set and make a preliminary recommendation. Here's how this person described his process.

I began by cleaning the data a little bit. In particular, I noticed that a handful of the buildings in the data set had very low occupancy rates - less than 10 percent of available space occupied. I decided to remove these buildings from consideration, on the theory that these buildings might have something weird going on with them, and could potentially distort the analysis. Once I scrubbed these low-occupancy buildings from the data set, I looked at the green buildings and non-green buildings separately. The median market rent in the non-green buildings was $25 per square foot per year, while the median market rent in the green buildings was $27.60 per square foot per year: about $2.60 more per square foot. I used the median rather than the mean, because there were still some outliers in the data, and the median is a lot more robust to outliers. Because our building would be 250,000 square feet, this would translate into an additional $250000 x 2.6 = $650000 of extra revenue per year if we build the green building.

Our expected baseline construction costs are $100 million, with a 5 percent expected premium for green certification. Thus we should expect to spend an extra $5 million on the green building. Based on the extra revenue we would make, we would recuperate these costs in $5000000/650000 = 7.7 years. Even if our occupancy rate were only 90 percent, we would still recuperate the costs in a little over 8 years. Thus from year 9 onwards, we would be making an extra $650,000 per year in profit. Since the building will be earning rents for 30 years or more, it seems like a good financial move to build the green building.

The developer listened to this recommendation, understood the analysis, and still felt unconvinced. She has therefore asked you to revisit the report, so that she can get a second opinion.

Do you agree with the conclusions of her on-staff stats guru? If so, point to evidence supporting his case. If not, explain specifically where and why the analysis goes wrong, and how it can be improved. Do you see the possibility of confounding variables for the relationship between rent and green status?
```python
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
%pylab inline
```

```python
df = pd.read_csv("greenbuildings.csv")
df[:5]
```

```python
df_cleaned = df[df['leasing_rate'] >= 0.1]

plt.figure(figsize=(10, 6))
df_cleaned.boxplot(column='Rent', by='green_rating', grid=False)
plt.title('Rent Distribution: Green vs Non-Green Buildings')
plt.suptitle('')
plt.xlabel('Green Building Status')
plt.ylabel('Rent (Per Square Foot)')
plt.xticks([1, 2], ['Non-Green', 'Green'])
plt.show()
```

When examining the conclusions of the developer's on-staff stats guru, one of his main claims was that the median market rent of the green buildings was greater than the median market rent of the non-green buildings. This is despite the number of outliers present in the data for each building type. To test this claim, I underwent the same steps as the guru did regarding cleaning the data. Any building with a leasing rate of less than 10% was disregarded from the data set, since such a low leasing rate can influence the rent per square foot signfiicantly, to the point where the data would not be reliable. After doing this, the above box plot was created, in order to offer a proper comparison between rent distributions for both green and non-green buildings. The x-axis represents the green building status for the building (either green or non-green), and the y-axis represents the rent per square foot of buidlings which fall under either green building status classification. From the box plot, it certainly seems that green buildings typically do have greater median rents per square foot relative to non-green buildings, albeit not by much. This supports the guru's claim of green buildings generally possessing higher rents. Granted, there is definitely some overlap when examining the rent distributions of both green building status groups, particularly due to the massive amount of outliers the non-green buildings have regarding rent per square foot. On this front, the guru's case is handily supported. However, as the next several figures indicate, there is a definite possibility of confounding variables for the relationship between rent and green status.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='size', y='Rent', hue='green_rating', data=df_cleaned)
plt.title('Rent vs Building Size (Green vs Non-Green Buildings)')
plt.xlabel('Building Size (Square Feet)')
plt.ylabel('Rent (Per Square Foot)')
plt.show()
```

One possible confounding variable when concerning the relationship between rent and green status relates to building size. No matter the status, there is the potential that a building is associated with a higher rent per square foot due to merely its square feet (size). To examine this further, a scatter plot was developed, with building size in square feet on the x-axis and rent per square foot on the y-axis. As indicated by the legend, the blue dots represent building with a green_rating of 0 (non-green building status), whereas the orange dots represent buildings with a green_rating of 1 (green building status). Given this information, it appears that the green buildings are usually a bit larger than the non-green buildings, with many of them hovering around the 1,000,000 square foot range. This perhaps suggests that a contributing factor for the green buildings possessing higher rents is a product of their size, not necessarily due to them being deemed better for the environment. In other words, an eco-friendly certification alone might not reliabily explain greater rents for building with green building status.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='leasing_rate', y='Rent', hue='green_rating', data=df_cleaned)
plt.title('Rent vs Leasing Rate (Green vs Non-Green Buildings)')
plt.xlabel('Leasing Rate (Occupancy)')
plt.ylabel('Rent (Per Square Foot)')
plt.show()
```

Another potential confounding variable in the relaitonship between rent and green status is the leasing rate, otherwise known as the occupancy rate for a building. Regardless of the eco-friendly status of a building, a greater leasing rate might correspond with a higher rent per sqaure foot of the building. If the building is a residential building, for instance, more tenants living in it would usually lead to more cumulative rent being paid, thereby hiking up the rent per square foot. To test this possible confounding nature, a scatter plot was created, with leasing rate on the x-axis and rent per sqaure foot on the y-axis. Similar to the previous scatter plot, the blue dots indicate non-green buildings, while the orange dots represent data from green buildings. From the plot, it can be observed that generally, higher leasing rates correlate with greater rents per square foot, for both non-green buildings and green buildings alike. That being said, the green buildings do appear to deviate towrds the higher rent per square foot values, irrespective of overall leasing rates. It is worth noting, though, that the non-green buildings behave in a somewhat similar fashion. Regardless, this trend among green buildings suggests that merely leasing rate might possibly not entirely explain the difference in rent between green and non-green buildings.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='Rent', hue='green_rating', data=df_cleaned)
plt.title('Rent vs Age (Green vs Non-Green Buildings)')
plt.xlabel('Building Age (Years)')
plt.ylabel('Rent (Per Square Foot)')
plt.show()
```

A third confounding variable in relation to the association between rent and green status is the age of the building. Newer buildings, no matter if they are eco-friendly or not, might command greater overall rents simply because of their modernity. Going back to the previous residential building example, an older building might not possess the infrastructure required to house large quantities of people, thereby lowering the rent per square foot of the property. Age, therefore, is very likely to play a substantial impact in relation to rent. To look for more insights into this potential phenomenon, a scatter plot was created, with the building age in years on the x-axis and the rent per square foot on the y-axis. Like the previous two scatter plots, the blue dots represent non-green building, while the orange dots represent green buildings. From the figure above it can be determined that green buildings do seem to be considerably newer than non-green buildings, which in turn possess higher rents per square foot. Non-green buildings, on the other hand, seem to have a greater variation in building ages, with a sizeable amount being more than 100 years old. As such, the building age definitely could be a confounding factor when it comes to the greater rent per sqaure foot observed in green buildings. In other words, the sustainability of the building might not be the only contributor to the higher rents exhibited in green buildings.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='class_a', y='Rent', hue='green_rating', data=df_cleaned)
plt.title('Rent vs Class A (Green vs Non-Green Buildings)')
plt.xlabel('Class A Building (1 = Yes, 0 = No)')
plt.ylabel('Rent (Per Square Foot)')
plt.show()
```

Yet another confounding variable when it comes to the relationship between rent and green status is whether or not the building is a Class A building. Because they are perceived to be of the highest quality, they might possess greater rents per sqaure foot, regardless of whether or not the building are of green status. It would make sense for a more valuable real estate building to demand a more significant rent payment per square foot, as it would cost more to maintain the building and account for other expenses it might have in the future. To further examine this possibility, a scatter plot was created, with whether or not the building is a Class A building on the x-axis (1 - Class A, 0 - not Class A) and rent per square foot on the y-axis. As was the case in the prior scatter plots, blue dots represent non-green buildings and orange dots represent green buildings. From the above scatter plot, there does seem to be a stark difference as it concerns rent per square foot between Class A and non-Class A buildings, with Class A buildings generally requiring more rent than their lower-quality counterparts. That being said, it is difficult to determine whether or not more green buildings are Class A relative to non-green buildings. If this is the case, though, the status of being Class A could be a massive confounder with regards to the higher rent demanded by green buildings.

It is a definite possibility that these factors can also slightly explain the difference in rent bewteen green and non-green buildings.
